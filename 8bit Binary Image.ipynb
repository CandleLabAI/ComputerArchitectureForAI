```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2

img = cv2.imread('test.jpeg')
```

```python
type(img)
```

```python
c = []
```

```python
c.append(img%2)
c.append((img//2)%2)
c.append((img//4)%2)
c.append((img//8)%2)
c.append((img//16)%2)
c.append((img//32)%2)
c.append((img//64)%2)
c.append((img//128)%2)
```

```python
cc = 16*c[4] + 32*c[5] + 64*c[6] + 128*c[7]
```

```python
plt.imshow(img)
plt.title("Original Image")
```

```python
plt.figure(figsize=(8,8))
for i in range(8):
    ax = plt.subplot(4, 2, i + 1)
    plt.imshow((2**i)*c[i])
    plt.title(i)
    plt.axis("off")
```

```python
plt.imshow(cc)
plt.title("Image with last 4 bits as 0")
```

```python
plt.imshow(img-cc)
plt.title("Difference between original image and image with last 4 bits as 0")
```
