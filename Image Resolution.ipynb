```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import cv2

img = cv2.imread('test.jpeg')
```

```python
plt.imshow(img)
plt.title("Original Image")
```

```python
type(img)
```

```python
x,y,z = np.shape(img)
```

```python
z
```

```python
img2 = np.zeros((x,y),dtype='int')
```

```python
for i in range(x):
    for j in range(y):
        img2[i][j] = (img[i][j][0]+img[i][j][1]+img[i][j][2])//3
```

```python
plt.imshow(img2)
plt.title("Original Image")
```

```python
img3 = np.zeros((x//2,y//2,z), dtype='int')
for i in range(x//2):
    for j in range(y//2):
        for k in range(z):
            img3[i][j][k] = img[2*i][2*j][k]
```

```python
np.shape(img3)
```

```python
ax = plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")
ax = plt.subplot(1, 2, 2)
plt.imshow(img3)
plt.title("Skipped pixel Image")
```
